```python
from sklearn.model_selection import train_test_split
```

```python
# Separar caracterÃ­sticas y la variable objetivo
X = df.drop(columns=['race_ethnicity'])
y = df['race_ethnicity']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Escalar las caracterÃ­sticas
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import BaggingClassifier
from sklearn.metrics import accuracy_score, classification_report
```

Bagging con KNeighborsClassifier

```python
model_knn = BaggingClassifier(
    estimator=KNeighborsClassifier(),
    n_estimators=100,
    random_state=42
).fit(X_train_scaled, y_train)
```

Prediccion y Evaluacion KNeighborsClassifier

```python
y_pred_knn = model_knn.predict(X_test_scaled)
accuracy_knn = accuracy_score(y_test, y_pred_knn)
report_knn = classification_report(y_test, y_pred_knn)

print("Accuracy con Bagging (KNN):", accuracy_knn)
print("Reporte de ClasificaciÃ³n con Bagging (KNN):\n", report_knn)
```

Bagging con DecisionTreeClassifier

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model_tree = BaggingClassifier(
    estimator=DecisionTreeClassifier(),
    n_estimators=100,
    random_state=42
).fit(X_train_scaled, y_train)
```

PredicciÃ³n y evaluaciÃ³n DecisionTreeClassifier

```python
y_pred_tree = model_tree.predict(X_test_scaled)
accuracy_tree = accuracy_score(y_test, y_pred_tree)
report_tree = classification_report(y_test, y_pred_tree)

print("Accuracy con Bagging (Decision Tree):", accuracy_tree)
print("Reporte de ClasificaciÃ³n con Bagging (Decision Tree):\n", report_tree)
```

ImplementaciÃ³n de Boosting (Gradient Boosting Classifier.)


```python
from sklearn.ensemble import GradientBoostingClassifier
```

```python
model_boost = GradientBoostingClassifier(
    n_estimators=100,
    random_state=42
).fit(X_train_scaled, y_train)
```

PredicciÃ³n y evaluaciÃ³n Boosting (Gradient Boosting Classifier.)

```python
y_pred_boost = model_boost.predict(X_test_scaled)
accuracy_boost = accuracy_score(y_test, y_pred_boost)
report_boost = classification_report(y_test, y_pred_boost)

print("Accuracy con Boosting (Gradient Boosting):", accuracy_boost)
print("Reporte de ClasificaciÃ³n con Boosting (Gradient Boosting):\n", report_boost)
```

```python
print("Accuracy con Bagging (KNN):", accuracy_knn)
print("Reporte de ClasificaciÃ³n con Bagging (KNN):\n", report_knn)

print("\nAccuracy con Bagging (Decision Tree):", accuracy_tree)
print("Reporte de ClasificaciÃ³n con Bagging (Decision Tree):\n", report_tree)

print("\nAccuracy con Boosting (Gradient Boosting):", accuracy_boost)
print("Reporte de ClasificaciÃ³n con Boosting (Gradient Boosting):\n", report_boost)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
from sklearn.metrics import classification_report, accuracy_score
```

```python
from sklearn.metrics import confusion_matrix
```

Entrenamiento con Modelo Random Forest

```python
modelo_rrff = RandomForestClassifier(n_estimators=100, random_state=42)
modelo_rrff.fit(X_train_scaled, y_train)

# Predecir en el conjunto de prueba
y_pred = modelo_rrff.predict(X_test_scaled)

# Evaluar el modelo
accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)
print(f'Accuracy: {accuracy}')
print(report)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_features': ['auto', 'sqrt', 'log2'],
    'max_depth': [4, 6, 8, 10],
    'criterion': ['gini', 'entropy']
}
```

```python
grid_search = GridSearchCV(estimator=modelo_rrff, param_grid=param_grid, cv=5, n_jobs=-1, verbose=2)
grid_search.fit(X_train_scaled, y_train)

# Mejor modelo encontrado
mejor_modelo_rrff = grid_search.best_estimator_

# Evaluar el mejor modelo
y_pred_best = mejor_modelo_rrff.predict(X_test_scaled)
best_accuracy = accuracy_score(y_test, y_pred_best)
best_report = classification_report(y_test, y_pred_best)
print(f'Best Accuracy: {best_accuracy}')
print(best_report)
```
